```python
import os
from dotenv import load_dotenv
import nest_asyncio
nest_asyncio.apply()


load_dotenv("../.env")

GEMINI_API_KEY= os.getenv("GEMINI_API_KEY")

if not GEMINI_API_KEY:
    print("API key not found! Please check your .env file.")
else:
    print("GEMINI API KEY successfully loaded!")
```

```python
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding
from llama_index.core import Settings

llm = GoogleGenAI(
    model="models/gemini-2.0-flash",
    api_key=GEMINI_API_KEY
)

embed_model = GoogleGenAIEmbedding(
    model_name="models/gemini-embedding-2-preview", 
    api_key=GEMINI_API_KEY
)

Settings.llm = llm
Settings.embed_model = embed_model

print("LlamaIndex successfully configured with Gemini 2.0 Flash!")
```

```python
from llama_index.core import SimpleDirectoryReader

documents= SimpleDirectoryReader("../data").load_data()
print(f"Loaded {len(documents)} document pages/sources!")

if documents:
    print(f"\nFirst Document Snippet: {documents[0].text[:200]}...")
else:
    print("\nData folder is empty! try adding a PDF or text file to the 'data/'.")
```

```python
from llama_index.core import VectorStoreIndex, StorageContext
from llama_index.core.node_parser import SentenceSplitter

node_parser= SentenceSplitter(chunk_size= 1024, chunk_overlap= 100)

nodes= node_parser.get_nodes_from_documents(documents)

index= VectorStoreIndex(nodes)

print(f"Created {len(nodes)} searchable Nodes!")
print("Baseline Index successfully built!")
```

```python
query_engine= index.as_query_engine(streaming= True)

user_query= "Summarize the key takeaways from this document."
response= query_engine.query(user_query)

print(f"Query: {user_query}")
print(f"Answer: ", end="")

for token in response.response_gen:
    print(token, end="", flush= True)
```
